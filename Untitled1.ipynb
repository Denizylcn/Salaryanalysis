```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python

dataset=r"C:\Users\win10\Desktop\verimadenciliği-işzekası\Salary.csv"
```

```python
veri = pd.read_csv(dataset, encoding='Windows-1254')
```

```python
sütunlar = veri.columns
print(sütunlar)
```

```python
veri
```

```python
bos_sutunlar = veri['Salary'].isnull().sum()

if bos_sutunlar > 0:
    print(f"Maaş sütununda {bos_sutunlar} boş değer bulunmaktadır.")
else:
    print("Maaş sütununda boş değer bulunmamaktadır.")
    
    bos_sutunlar = veri['Education Level'].isnull().sum()

if bos_sutunlar > 0:
    print(f"Education Level sütununda {bos_sutunlar} boş değer bulunmaktadır.")
else:
    print("Education Level sütununda boş değer bulunmamaktadır.")
```

```python
plt.figure(figsize=(10, 6))
veri.boxplot(column='Salary', by='Education Level', vert=True)
plt.ylabel('Maaş')
plt.xlabel('Eğitim Seviyesi')
plt.title('Maaş ve Eğitim Seviyesine Göre Box Plot')
# Grafiği belirtilen dizine kaydedin
plt.savefig(r'C:\Users\win10\Desktop\verimadenciliği-işzekası\salary-education')

plt.show()

```

```python
plt.figure(figsize=(10, 6))
veri.boxplot(column='Age', by='Education Level', vert=True)
plt.ylabel('Yaş')
plt.xlabel('Eğitim Seviyesi')
plt.title('Yaş ve Eğitim Seviyesine Göre Box Plot')
plt.savefig(r'C:\Users\win10\Desktop\verimadenciliği-işzekası\age-education')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
veri.boxplot(column='Education Level', by='Gender', vert=True)
plt.xlabel('Cinsiyet')
plt.ylabel('Eğitim Seviyesi')
plt.title('Cinsiyet ve Eğitim Seviyesine Göre Box Plot')
plt.savefig(r'C:\Users\win10\Desktop\verimadenciliği-işzekası\gender-education')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
veri.boxplot(column='Salary', by='Country', vert=True)
plt.ylabel('Maaş')
plt.xlabel('Ülke')
plt.title('Maaş ve Ülkelere Göre Box Plot')
plt.savefig(r'C:\Users\win10\Desktop\verimadenciliği-işzekası\salary-country')
plt.show()

```

```python
# Veriyi gruplayın ve toplamı hesaplayın
gruplanmış_veri = veri['Education Level'].value_counts()

# Pasta grafiğini çizin
plt.figure(figsize=(8, 8))
plt.pie(gruplanmış_veri, labels=gruplanmış_veri.index, autopct='%1.3f%%', startangle=90)
plt.title('Eğitim Seviyelerinin Dağılımı')

# Toplam kişi sayısını hesaplayın
toplam_kisi_sayisi = gruplanmış_veri.sum()

# Grafiği gösterin
plt.show()

# Toplam kişi sayısını yazdırın
print(f"Toplam Kişi Sayısı: {toplam_kisi_sayisi}")
```

```python
gruplanmış_veri = veri['Gender'].value_counts()

# Pasta grafiğini çizin
plt.figure(figsize=(8, 8))
plt.pie(gruplanmış_veri, labels=gruplanmış_veri.index, autopct='%1.1f%%', startangle=90)
plt.title('Cinsiyet Dağılımı')
toplam_kisi_sayisi = gruplanmış_veri.sum()
# Grafiği gösterin
plt.show()
print(f"Toplam Kişi Sayısı: {toplam_kisi_sayisi}")
```
